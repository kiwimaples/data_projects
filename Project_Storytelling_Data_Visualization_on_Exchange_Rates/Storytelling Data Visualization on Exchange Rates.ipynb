# Storytelling Data Visualization on Exchange Rates

We will be using data visualisation techniques to tell a story in the Euro daily exchange rates betwen 1999 to 2021, mainly in relations to the US Dollar.

The euro (symbolized with €) is the official currency in most of the countries of the European Union.

The data source was from the European Central Bank and has been made available [here](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020)

```python
# Remove comments from Jupyter
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

# Reading the dataset
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')

# First 5 rows
exchange_rates.head()
```

```python
# Last 5 rows
exchange_rates.tail()
```

Dataset starts with the most recent data first and goes towards the first recorded date to the end

```python
# Further info on the dataset
exchange_rates.info()
```

From observation:
* There are 5699 rows and 41 columns
* There are a good amount of columns with null values, most if not all being soft/exotic currencies
* The main currencies from developed countries (US Dollar etc) do not have any empty values
* There are 3 numerical (float) data and 31 text (object) data

# Data Cleaning

```python
# Renaming columns
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar', 'Period\\Unit:': 'Time'}, inplace=True) 
```

```python
exchange_rates.columns
```

```python
# Convert Time column to date and time format
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
```

```python
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates.head()
```

```python
exchange_rates.tail()
```

Order of the rows have been change to start with the start date of the dataset

```python
# Isolate to only columns we will work with
euro_to_dollar = exchange_rates[['Time','US_dollar']]
euro_to_dollar.head()
```

```python
# Creating a frequency table
euro_to_dollar['US_dollar'].value_counts()
```

There are 62 rows where the Time value is '-'. This will be removed.

```python
# Remove rows containing '-'
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']

# Convert USD to numerical values
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

# Rolling Mean/Moving Average

```python
# Creating charts for visualisation
import matplotlib.pyplot as plt

# Allows Jupyter to display graphs
%matplotlib inline 
```

```python
# Creating a column with data containing the 30 day moving average
euro_to_dollar['30MA'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

# Story Ideas

Here are a few story ideas for our data:

* We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 to the most recent data and the 2016-2019 data as a baseline. We can use a line plot.
* We show how the euro-dollar rate changed during the 2007-2008's financial crisis. We can also show the data for 2006 and 2009 for comparison. We can use a line plot.
* We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

# Storytelling Data Visualization Examples

# Coronavirus Pandemic Example

```python
# Data from 2016 until the covid period
pre_covid = euro_to_dollar.copy(
)[(euro_to_dollar['Time'].dt.year >= 2016) & (euro_to_dollar['Time'].dt.year <= 2019)]

# Data representing the covid period
covid = euro_to_dollar.copy()[euro_to_dollar['Time'].dt.year >= 2020]
```

```python
# Statistical information for the 30 day moving average to identify the top and bottom of the graph
covid['30MA'].describe()
```

```python
# Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(8,4))

# Pre-Covid period plot
ax.plot(pre_covid['Time'], pre_covid['30MA'], color='blue', alpha=0.5)
ax.set_xticklabels(['', '2016', '2017', '2018','2019', '2020', '2021'], alpha=0.3)
ax.set_yticklabels(['','1.05','1.1','1.15','1.2','1.25'], alpha=0.3)

# Focus point
ax.axvspan(xmin=18250, xmax=18420, ymin=0.03, facecolor='grey', alpha=0.2)

# Covid period Plot
ax.plot(covid['Time'], covid['30MA'], color='red')

# Title area
ax.text(16580,1.27,'Euro peaked at 1.22 during the Coronavirus Pandemic', fontsize=18, weight='bold')
ax.text(16580,1.255,'Euro-USD exchange rates before Covid and during Covid', fontsize=14)

# Adding signature
ax.text(16570, 1, '' + ' '*145 + 'Source: European Central Bank',
        color = 'white', backgroundcolor = 'black',
        size=9)

plt.show()
```

## Financial Crisis Example

```python
financial_crisis = euro_to_dollar.copy(
                    )[(euro_to_dollar['Time'].dt.year >= 2006) & (euro_to_dollar['Time'].dt.year <= 2009)] 

financial_crisis_7_8 = euro_to_dollar.copy(
                    )[(euro_to_dollar['Time'].dt.year >= 2007) & (euro_to_dollar['Time'].dt.year <= 2008)] 
```

```python
# Adding the FiveThirtyEight style
style.use('fivethirtyeight')

# Adding the plot
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(financial_crisis['Time'], financial_crisis['30MA'], linewidth=1, color='green')
ax.plot(financial_crisis_7_8['Time'], financial_crisis_7_8['30MA'], linewidth=3, color='red')

# Highlighting the peak of the crisis
ax.axvspan(xmin=13950, xmax=14150, ymin=0.08, facecolor='grey', alpha=0.2)

# Adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 13105
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=11)
    x += 365

y= 1.195
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(13000, y, rate, alpha=0.5, fontsize=11)
    y += 0.1
    
ax.text(13000,1.66,"Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis", weight='bold', fontsize=12)
ax.text(13000,1.62,"Euro-USD exchange rates between 2006 and 2010", fontsize=11)

# Adding signature
ax.text(13000, 1.05, '' + ' '*113 + 'Source: European Central Bank',
        color = 'white', backgroundcolor = 'black',
        size=9)

plt.show()
```

# The Three US Presidencies Example

```python
# Period for all 3 presidents
bush_obama_trump = euro_to_dollar.copy(
                    )[(euro_to_dollar['Time'].dt.year >=2001) 
                      & (euro_to_dollar['Time'].dt.year <=2021)]

# Bush presidency period
bush = euro_to_dollar.copy(
        )[(euro_to_dollar['Time'].dt.year >=2001) 
          & (euro_to_dollar['Time'].dt.year < 2009)]

# Obama presidency period
obama = euro_to_dollar.copy(
        )[(euro_to_dollar['Time'].dt.year >=2009) 
          & (euro_to_dollar['Time'].dt.year < 2017)]

# Trump presidency period
trump = euro_to_dollar.copy(
        )[(euro_to_dollar['Time'].dt.year >=2017) 
          & (euro_to_dollar['Time'].dt.year < 2021)]
```

```python
style.use('fivethirtyeight')

plt.figure(figsize=(12,6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

for ax in axes:
    ax.set_ylim(0.8,1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.grid(alpha=0.5)

# Bush
ax1.plot(bush['Time'], bush['30MA'], color='#BF5FFF')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '','2007', '', '2009'],alpha=0.3)
ax1.set_yticklabels(['1.0', '1.2', '1.4', '1.6'],alpha=0.3)
ax1.text(12500,1.88, "BUSH", color='#BF5FFF', weight='bold', fontsize=18)
ax1.text(12200,1.78, "(2001-2009)", alpha=0.3, weight='bold')

# Obama    
ax2.plot(obama['Time'], obama['30MA'], color='#ffa500')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '','2015', '', '2017'],alpha=0.3)
ax2.set_yticklabels(['1.0', '1.2', '1.4', '1.6'],alpha=0.3)
ax2.text(15150,1.88, "OBAMA", color='#ffa500', weight='bold', fontsize=18)
ax2.text(14990,1.78, "(2008-2017)", alpha=0.3,  weight='bold')

# Trump
ax3.plot(trump['Time'], trump['30MA'], color='#00B2EE')
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '', '2020','', '2021', ''],alpha=0.3)
ax3.set_yticklabels(['1.0', '1.2', '1.4', '1.6'],alpha=0.3)
ax3.text(17700,1.88, "TRUMP", color='#00B2EE', weight='bold', fontsize=18)
ax3.text(17600,1.78, "(2017-2021)", alpha=0.3, weight='bold')

# All Ppresident years
ax4.plot(bush_obama_trump['Time'], bush_obama_trump['30MA'])
ax4.plot(bush['Time'], bush['30MA'], color='#BF5FFF')
ax4.plot(obama['Time'], obama['30MA'], color='#ffa500')
ax4.plot(trump['Time'], trump['30MA'], color='#00B2EE')

# Title text
ax1.text(10800,2.30, '''EURO-USD rate averaged 1.22 under the last three US presidents''', fontsize=20, weight='bold')

ax1.text(10800,2.10, '''EURO-USD exchange rates under George W. Bush (2001 - 2009),Barack Obama (2009-2017),
and Donald Trump (2017-2021)''', fontsize=16)

# Adding signature
ax.text(10700, 0.7, '' + ' '*125 + 'Source: European Central Bank',
        color = 'white', backgroundcolor = 'black',
        size=14)

plt.show()
```
